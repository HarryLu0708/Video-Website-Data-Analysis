```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv("PornLens_Pornstars_Data.csv")
data.head()
```

```python
data.fillna("Missing")
data.head()
```

```python
data.isnull().mean()
```

```python
drop_columns = ["Unnamed: 0.1","Unnamed: 0","Facial Hair","Interests and hobbie","Hometown","Foreskin","Endowment","Turn Ons","Turn Offs","Eye Color","Background","Measurements","Star Sign","Career Start and End"]
```

```python
dropped_data = data.drop(drop_columns,axis=1)
dropped_data.head()
```

```python
dropped_data = dropped_data.drop(["Links","Names","Weekly Ranks","Monthly Rank","Last Month Rank","Yearly Rank"],axis=1)
dropped_data.isnull().mean()
```

```python
dropped_data = dropped_data.drop(["Images","Social Medias","Interests and hobbies"],axis=1)
dropped_data.isnull().mean()
```

```python
dropped_data.dropna(subset=["City and Country","Ethnicity","Fake Boobs"],inplace=True)
dropped_data.isnull().mean()
```

```python
def null_return(val):
    if val=="missing":
        return False
    return True
```

```python
dropped_data["Featured In"] = dropped_data["Featured In"].fillna("missing")
dropped_data["Featured In"] = dropped_data["Featured In"].apply(null_return)
dropped_data["Featured In"]
dropped_data["Featured In"].unique()
```

```python
dropped_data["Featured In"]
```

```python
dropped_data["Other"] = dropped_data["Other"].fillna("missing")
dropped_data["Other"] = dropped_data["Other"].apply(null_return)

dropped_data["Official Site"] = dropped_data["Official Site"].fillna("missing")
dropped_data["Official Site"] = dropped_data["Official Site"].apply(null_return)
```

```python
def data_return(dropped_data,name):
    dropped_data[name] = dropped_data[name].fillna("missing")
    dropped_data[name] = dropped_data[name].apply(null_return)
    return dropped_data
```

```python
dropped_data = data_return(dropped_data,"Instagram")
dropped_data = data_return(dropped_data,"Twitter")
dropped_data = data_return(dropped_data,"Snapchat")
dropped_data = data_return(dropped_data,"Amazon Wishlist")
dropped_data.isnull().mean()
```

```python
dropped_data = data_return(dropped_data,"Abouts")
dropped_data.isnull().mean()
```
